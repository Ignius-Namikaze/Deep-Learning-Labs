```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

# Sigmoid Function
f(x) = 1/(1+e^-x)

```python
def sigmoid(x):
  return 1/(1+ np.exp(-x))

x = np.linspace(-10,10,100)
y = sigmoid(x)

plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Sigmoid of X")
plt.title("Sigmoid activation Function")
plt.grid(True)
plt.show()
```

f'(x) = f(x)(1-f(x))

```python
x = np.linspace(-10,10,100)
y = sigmoid(x)*(1-sigmoid(x))

plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Sigmoid of X")
plt.title("Derivative of Sigmoid activation Function")
plt.grid(True)
plt.show()
```

# TanH Function
f(x) = tanh(x) = 2/1 + e^-2x

```python
def tanh(x):
  return (2/(1+ np.exp(-2*x)))-1

x = np.linspace(-10,10,200)
y = tanh(x)

plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("TanH of X")
plt.title("TanH activation Function")
plt.grid(True)
plt.show()
```

f'(x) = 1 - tanH(x)^2

```python
x = np.linspace(-10,10,200)
y = 1-(tanh(x)*tanh(x))

plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("TanH of X")
plt.title("Derivative of TanH activation Function")
plt.grid(True)
plt.show()
```

# ReLU Function
f(x) = 0 for x<0,
f(x) = x for x>=0

```python
def relu(x):
    if x < 0:
        return 0
    else:
        return x

x = np.linspace(-10,10,400)
y = np.vectorize(relu)(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='ReLU Activation', color='blue')
plt.axhline(0, color='red', linestyle='--', linewidth=0.5)
plt.axvline(0, color='green', linestyle='--', linewidth=0.5)
plt.xlabel('Input (x)')
plt.ylabel('Output (ReLU(x))')
plt.title('ReLU Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```

f'(x) = 0 for x<0,
f'(x) = 1 for x>=0

```python
def relu_derivative(x):
    if x < 0:
        return 0
    else:
        return 1

x = np.linspace(-2, 2, 400)
y_derivative = np.vectorize(relu_derivative)(x)

plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 2)
plt.plot(x, y_derivative, label='ReLU Derivative', color='purple')
plt.axhline(0, color='red', linestyle='--', linewidth=0.5)
plt.axvline(0, color='green', linestyle='--', linewidth=0.5)
plt.xlabel('Input (x)')
plt.ylabel('Derivative (dReLU(x)/dx)')
plt.title('Derivative of ReLU Activation Function')
plt.legend()
plt.grid(True)
plt.show()
```
