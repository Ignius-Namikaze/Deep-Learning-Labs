```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
df_1 = pd.get_dummies(df, columns = ['Species'])
```

```python
df_1.head(5)
```

```python
df.drop("Species", axis=1, inplace=True)
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
df_scaled = std_scaler.fit_transform(df)
```

```python
df_transformed = pd.DataFrame(df_scaled, columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
```

```python
df_transformed.head(5)
```

```python
df_1.drop(columns=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"], inplace=True)
```

```python
df_1
```

```python
Iris_df=pd.concat([df_transformed,df_1],axis=1)
```

```python
input = Iris_df.iloc[:,0:4]
```

```python
input.shape
```

```python
target = Iris_df.iloc[:,4:7]
```

```python
target.shape
```

## MLP

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(input,target,test_size=0.2, random_state=42)
```

```python
X_train_mlp,X_val_mlp,Y_train_mlp,Y_val_mlp=train_test_split(X_train,Y_train,test_size=0.2)
```

```python
print(X_train_mlp.shape)
```

```python
print(X_val_mlp.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
from keras.datasets import mnist
#from keras.utils import np_utils
```

```python
batch_size=10
num_epochs=50
```

```python
model=Sequential()
model.add(Dense(input_dim=4,activation='sigmoid',units=3,kernel_initializer='normal'))
model.add(Dense(input_dim=3,activation='softmax',units=3,kernel_initializer='normal'))
model.compile(optimizer=SGD(learning_rate=0.1),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history=model.fit(X_train_mlp, Y_train_mlp,
                  batch_size=batch_size,
                  epochs=num_epochs,
                  verbose=1,
                  validation_data=(X_val_mlp,Y_val_mlp))
```

```python
score=model.evaluate(X_test,Y_test,verbose=1)
print()
print('MLP Test loss: ',score[0])
print('MLP Test Accuracy: ',score[1])
```

## Changing learning rate

```python
model2=Sequential()
model2.add(Dense(input_dim=4,activation='sigmoid',units=3,kernel_initializer='normal'))
model2.add(Dense(input_dim=3,activation='softmax',units=3,kernel_initializer='normal'))
model2.compile(optimizer=SGD(learning_rate=0.001),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history2=model2.fit(X_train_mlp, Y_train_mlp,
                  batch_size=batch_size,
                  epochs=num_epochs,
                  verbose=1,
                  validation_data=(X_val_mlp,Y_val_mlp))
```

## Therefore we conclude that learning rate=0.1 gave us the best result


## Adding momentum

```python
model3=Sequential()
model3.add(Dense(input_dim=4,activation='sigmoid',units=3,kernel_initializer='normal'))
model3.add(Dense(input_dim=3,activation='softmax',units=3,kernel_initializer='normal'))
model3.compile(optimizer=SGD(learning_rate=0.1,momentum=0.5),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history4=model3.fit(X_train_mlp, Y_train_mlp,
                  batch_size=batch_size,
                  epochs=num_epochs,
                  verbose=1,
                  validation_data=(X_val_mlp,Y_val_mlp))
```

```python
score3=model3.evaluate(X_test,Y_test,verbose=1)
print()
print('MLP Test loss: ',score3[0])
print('MLP Test Accuracy: ',score3[1])
```

```python
model4=Sequential()
model4.add(Dense(input_dim=4,activation='sigmoid',units=3,kernel_initializer='normal'))
model4.add(Dense(input_dim=3,activation='softmax',units=3,kernel_initializer='normal'))
model4.compile(optimizer=SGD(learning_rate=0.1,momentum=0.9),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
history4=model4.fit(X_train_mlp, Y_train_mlp,
                  batch_size=batch_size,
                  epochs=num_epochs,
                  verbose=1,
                  validation_data=(X_val_mlp,Y_val_mlp))
```

```python
score4=model4.evaluate(X_test,Y_test,verbose=1)
print()
print('MLP Test loss: ',score4[0])
print('MLP Test Accuracy: ',score4[1])
```
